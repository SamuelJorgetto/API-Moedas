```python
import pandas as pd
import requests
from datetime import datetime
from sqlalchemy import create_engine

moedas = ['USD-BRL', 'EUR-BRL', 'BTC-BRL', 'ARS-BRL']

url = f"https://economia.awesomeapi.com.br/json/last/{','.join(moedas)}"
response = requests.get(url)
data = response.json()

dados_formatados = []
for par in moedas:
    key = par.replace('-', '')
    info = data[key]
    dados_formatados.append({
        'moeda': info['code'],
        'cotacao': float(info['bid']),
        'alta': float(info['high']),
        'baixa': float(info['low']),
        'data_hora': datetime.fromtimestamp(int(info['timestamp']) / 1000)
    })

df = pd.DataFrame(dados_formatados)

display(df)

engine = create_engine('sqlite:///cotacoes.db')
df.to_sql('cotacoes_moedas', con=engine, if_exists='append', index=False)

```
